```python
import pandas as pd
```

```python
def format_data(data: pd.DataFrame) -> pd.DataFrame:
    data = data.drop(['urlDocumento', 'numSubCota', 'numEspecificacaoSubCota', 'txtNumero', 'numLote', 'nuCarteiraParlamentar', 'codLegislatura', 'ideDocumento'], axis=1)
    return data
```

```python
data_2022 = pd.read_csv('./data/Ano-2022.csv', sep=';')
data_2021 = pd.read_csv('./data/Ano-2021.csv', sep=';')
data_2020 = pd.read_csv('./data/Ano-2020.csv', sep=';')
data_2019 = pd.read_csv('./data/Ano-2019.csv', sep=';')

frames = [data_2022, data_2021, data_2020, data_2019]

data = pd.concat(frames)
```

```python
print(data.columns)
```

```python
formated_data = format_data(data)
print(formated_data.head(5))
```

```python
expenses_by_txNomeParlamentar = data.groupby('txNomeParlamentar').sum().sort_values(by='vlrDocumento', ascending=False)
print(expenses_by_txNomeParlamentar.head(10)['vlrDocumento'])
```

```python
expenses_by_sgPartido = data.groupby('sgPartido').sum().sort_values(by='vlrDocumento', ascending=False)
print(expenses_by_sgPartido.head(10)['vlrDocumento'])
```

```python
expenses_by_txtFornecedor = data.groupby('txtFornecedor').sum().sort_values(by='vlrDocumento', ascending=False)
print(expenses_by_txtFornecedor.head(20)['vlrDocumento'])
```

```python
expenses_by_txtDescricao = data.groupby('txtDescricao').sum().sort_values(by='vlrDocumento', ascending=False)
print(expenses_by_txtDescricao.head(20)['vlrDocumento'])
```
